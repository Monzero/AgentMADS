```python
def load_and_preprocess(pdf_path):
    """Load and preprocess the annual report PDF."""
    loader = PyPDFLoader(pdf_path)
    documents = loader.load_and_split()
    return documents
```

```python
class SentenceTransformerEmbedding(Embeddings):
    def __init__(self, model_name: str = "all-MiniLM-L6-v2"):
        self.model = SentenceTransformer(model_name)

    def embed_documents(self, texts):
        """Embed a list of documents."""
        return self.model.encode(texts, convert_to_tensor=False, show_progress_bar=True)

    def embed_query(self, text):
        """Embed a single query."""
        return self.model.encode(text, convert_to_tensor=False, show_progress_bar=False)
```

```python
def create_vector_store(documents):
    """Generate embeddings using SentenceTransformers and create a vector store."""
    embedding_model = SentenceTransformerEmbedding()  # Use the custom embedding class
    texts = [doc.page_content for doc in documents]
    vector_store = FAISS.from_texts(texts, embedding=embedding_model)
    return vector_store
```

```python
serpapi_api_key = "a03af5c777130fb07b5c3cf8e58e0731c23bd3367ddd15d15bfab081249687f4"
search_tool = SerpAPIWrapper(serpapi_api_key=serpapi_api_key)
```

```python
retriever = vector_store.as_retriever()
rag_chain = RetrievalQA.from_chain_type(
    llm=llm,
    chain_type="stuff",  # 'stuff', 'map_reduce', or 'refine' based on preference
    retriever=retriever
)
```

```python
def document_or_web_search(query):
    response = rag_chain.run(query)
    if response and "I don't know" not in response:
        return response  # Return from vector DB if found
    return search_tool.run(query)  # Otherwise, search the web
```

```python
tools = [
    Tool(
        name="AnnualReportSearch",
        func=document_or_web_search,
        description="Search the company's annual report for answers. If not found, searches the web."
    )
]
```

```python
react_prompt = PromptTemplate(
    template="""You are an AI assistant that follows the ReAct format strictly.
Answer user questions using available tools if needed.

Format:
Thought: Think step-by-step about the question.
Action: Call the necessary tool if required (e.g., "AnnualReportSearch").
Action Input: Provide the required input for the tool.
Observation: Use the result to generate the final answer.

Example:
User: What was the company's net profit last year?
Thought: I should search the annual report for financial figures.
Action: AnnualReportSearch
Action Input: "Net profit last year"
Observation: [Tool response]
Thought: Now I can answer based on the document.
Final Answer: [Company's net profit]

Now, answer the user's question.

User: {input}
""",
    input_variables=["input"]
)
```
